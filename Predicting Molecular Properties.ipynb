```python
import ase
import pandas as pd
import numpy as np
from ase import Atoms
import ase.visualize
import random
```

```python
structures_df = pd.read_csv('./structures.csv')
```

```python
structures_df.head()
```

Let's visualize a random molecule

```python
random_molecule = random.choice(structures_df['molecule_name'].unique())
struct_1 = structures_df[structures_df['molecule_name'] == random_molecule]
atoms_struct_1 = struct_1[['x','y','z']]
symbols_struct_1=struct_1['atom']
system = Atoms(positions=atoms_struct_1, symbols=symbols_struct_1)
ase.visualize.view(system, viewer="x3d")
```

```python
import matplotlib.pyplot as plt
from ase.visualize.plot import plot_atoms
from ase.lattice.cubic import FaceCenteredCubic
slab = FaceCenteredCubic('Au', size=(2, 2, 2))
fig, ax = plt.subplots()
plot_atoms(system, ax, radii=0.3, rotation=('10x,145y,0z'))
plt.show()
```
